```python
from sklearn import linear_model
import matplotlib.pyplot as plt
import math

df_x, df_y = df[["PovPct"]], df["Brth15to17"]

model = linear_model.LinearRegression()
model.fit(df_x, df_y)

x_test = np.linspace(5, 25)
y_test = model.predict(x_test[:,None])

plt.scatter(df_x, df_y)
plt.plot(x_test, y_test)
plt.xlabel("Poverty %")
plt.ylabel("Birth Rate 15-17")
plt.show()

print("Intercept: %s" %model.intercept_)
print("Coefficient: %s" %model.coef_)
# error term is mean squared error
print("Error Term: %s" %math.sqrt(model._residues/len(df_y)))
```
